```python
import numpy as np
import pandas as pd
import re
import json
from konlpy.tag import Okt
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
f = open('./data_in/ChatBotData.csv','r', encoding='utf-8')
chatBotData = [line.strip().split(',') for line in f.readlines()]
```

```python
chatBotData[1:]
```

```python
okt = Okt()
voca = []
for line in chatBotData[1:]:
    for text in line[:2]:
        voca.extend(okt.morphs(text))
```
